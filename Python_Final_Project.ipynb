```python
import requests
from bs4 import BeautifulSoup
from selenium import webdriver
from selenium.webdriver.support.ui import Select
import time
```

```python
job_104_url_ECE = "https://www.104.com.tw/jobs/search/?ro=0&jobcat=2008001020%2C2008001016%2C2008001014&expansionType=area%2Cspec%2Ccom%2Cjob%2Cwf%2Cwktm&area=6001006000&order=11&asc=0&page=8&mode=s&jobsource=tab_cs_to_job" 
```

```python
browser = webdriver.Chrome()

browser.get(job_104_url_ECE)

time.sleep(2)  ## delay 一段時間等網頁回應再解析，否則會當掉
for i in range(10):
    time.sleep(2)
    browser.execute_script("window.scrollTo(0,10000);")
    
    
html = browser.page_source

soup = BeautifulSoup(html, 'lxml')

company_list = soup.find("div", attrs={'id':"js-job-content"}).find_all('article')
#print(company_list) 
for d in company_list:
    print("----------------------------------------------------------------------")
    #職稱
    job_title = d.a.text
    print(job_title)
    # 公司名稱
    company_name = d.find_all('li')[1].a['title']
    print(company_name)
    # 業務類別
    job_class = "職務類別:"+d.find_all('li')[2].text
    print(job_class)
    # 網址
    job_link = "104應徵網站:" + " " + "https:"+ d.a['href']
    print(job_link)
    # 工作內容
    job_content = d.p.text
    print(job_content)
    # 薪資
    job_salary = "薪資:" + d.find('span' , attrs = {'class' , 'b-tag--default'}).text
    print(job_salary)
```
